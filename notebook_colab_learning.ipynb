```python
import json
import numpy as np

for b in [1, 5, 50, 500, 5000]:
    data = json.load(open(f"MLP_lr0.01_b{b}.json"))
    best_epoch = int(np.argmax(data["val"]) + 1)
    test_acc = data["test"]
    updates = (10000 // b) * best_epoch
    print(f"batch={b}, best_epoch={best_epoch}, gradient_updates={updates}, test_acc={test_acc}")
```
